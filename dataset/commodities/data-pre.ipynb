```python
import pandas as pd
```

```python
meta = pd.read_csv("./_meta.txt")
meta
```

```python
csv_list = []
for symbol in meta["Symbol"]:
    filename = symbol + ".csv"
    df = pd.read_csv(filename)
    df['Symbol'] = symbol
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    csv_list.append(df)
```

```python

vertical_stack = None
for df in csv_list:
    if vertical_stack is None:
        vertical_stack = csv_list[0]
    else:
        vertical_stack = pd.concat([vertical_stack, df], axis=0)
vertical_stack
```

```python
vertical_stack.to_csv("_commo_merged.csv", index=False)
```
