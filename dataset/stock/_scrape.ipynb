```python
# NOTE: Country code use International Olympic Countries/Regions three letter code
# See details in https://www.wiki.zh-cn.nina.az/%E5%9B%BD%E9%99%85%E5%A5%A5%E5%A7%94%E4%BC%9A%E5%9B%BD%E5%AE%B6%E7%BC%96%E7%A0%81.html
df = pd.read_csv("_metadata.txt")
df
```

```python
# store historical data into each csv files relatively
for i in df.index:
    curr = yf.Ticker(df["Symbol"][i])
    hist = curr.history(period="max", actions=False)
    #hist.to_csv('{}.csv'.format(df["Continent"][i]+"_"+df["Name"][i]))
    hist.to_csv('{}.csv'.format( df["Name"][i]))
```

```python
# NOTE: commented lines (non-comment) for possible use

# add symbol & name cols
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))

# loop over the list of csv files
for f in csv_files:
    # read the csv file
    df_csv = pd.read_csv(f)
    
    # add Names & Symbol cols to csv
    name = f.split(sep='\\')[7].split(sep = '.')[0]
    symbol = df.loc[df["Name"] == name, "Symbol"].to_string(index=None)
    # continent = df.loc[df["Name"] == name, "Continent"].to_string(index=None)
    # country = df.loc[df["Name"] == name, "Country"].to_string(index=None)
    
    df_csv["Name"] = name
    df_csv["Symbol"] = symbol
    # df_csv["Continent"] = continent
    # df_csv["Country"] = country
    
    #store back into csv file
    df_csv.to_csv(f, index = None)
```

```python
# merge all csv to one

# joining files with concat and read_csv
df = pd.concat(map(pd.read_csv, csv_files), ignore_index=True)
df.to_csv("_stock_merged.csv", index=None)
```
